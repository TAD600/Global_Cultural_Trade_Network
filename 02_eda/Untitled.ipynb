```python
persistence = {}

for i in range(1, len(year_cols)):
    
    prev_year = year_cols[i-1]
    curr_year = year_cols[i]
    
    persisted = (
        (df_binary[prev_year] == 1) &
        (df_binary[curr_year] == 1)
    ).sum()
    
    persistence[curr_year] = persisted

persistence_df = pd.DataFrame({
    "year": persistence.keys(),
    "persistent_ties": persistence.values()
})

print(persistence_df)
```

```python
new_ties = {}

for i in range(1, len(year_cols)):
    
    prev_year = year_cols[i-1]
    curr_year = year_cols[i]
    
    formed = (
        (df_binary[prev_year] == 0) &
        (df_binary[curr_year] == 1)
    ).sum()
    
    new_ties[curr_year] = formed

new_ties_df = pd.DataFrame({
    "year": new_ties.keys(),
    "new_ties": new_ties.values()
})

print(new_ties_df)
```

```python
survival_rate = {}

for i in range(1, len(year_cols)):
    
    prev_year = year_cols[i-1]
    curr_year = year_cols[i]
    
    prev_total = (df_binary[prev_year] == 1).sum()
    
    persisted = (
        (df_binary[prev_year] == 1) &
        (df_binary[curr_year] == 1)
    ).sum()
    
    survival_rate[curr_year] = persisted / prev_total

survival_df = pd.DataFrame({
    "year": survival_rate.keys(),
    "survival_rate": survival_rate.values()
})
```
